```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
#1a
x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 3, 2, 4, 4])
y_pred = 2 * x + 1
MSE_a = np.square(np.subtract(y,y_pred)).mean()
print (MSE_a)

a = float(input("nhap gia tri a: "))
b = float(input("nhap gia tri b: "))
y_pred_b = a * x + b
MSE_b = np.square(np.subtract(y,y_pred_b)).mean()
print (MSE_b)

x = x.reshape(-1, 1)
reg = LinearRegression().fit(x, y)
print("Intercept: ", reg.intercept_)
print("Coefficient: ", reg.coef_)
print("Score: ", reg.score(x, y))
MSE_c = np.square(np.subtract(y,reg.predict(x))).mean()
print (MSE_c)
```

```python
#2
data = pd.read_csv(r'C:\555\baitap\LinearRegression\data\data_linear.csv')
data.columns = ['Area','Price']

x = data["Area"]
y = data["Price"]
plt.scatter(x, y)

x = x.values.reshape(-1,1)
reg = LinearRegression().fit(x, y)
print("Coefficient: ", reg.coef_)
MSE_2 = np.square(np.subtract(y, reg.predict(x))).mean()
print ("MSE 2: ", MSE_2)

print("Gia nha 50, 120, 150 (mÂ²): ", reg.predict([[50]]), reg.predict([[120]]), reg.predict([[150]]))

print("Score: ", reg.score(x, y))
plt.plot(x, reg.predict(x))
```
